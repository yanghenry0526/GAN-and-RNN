```python
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.layers import Dense,Dropout,Input
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.optimizers import Adam
from PIL import Image
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
import os


def load_data():
    x_train = []
    y_train = []
    childen_train_path = "C:/Users/asus/childen_saved/"
    childen_train_filedir = os.listdir(childen_train_path)
    for file in childen_train_filedir :
        im = Image.open(childen_train_path + file)
        data = np.array(im)
        x_train.append(data)
        y_train.append(0)
    
    x_train = np.array(x_train)
    x_train = (x_train.astype(np.float32)/255)
    x_train = x_train.reshape(10000,56*56)
    
    return(x_train,y_train)

x_train,y_train = load_data()
print(x_train[0].shape)



def build_generator(): 
    model = Sequential()
    
    model.add(Dense(units=128, input_dim=3136))   #100次輸入
    model.add(LeakyReLU(alpha=0.2))
    
    model.add(Dense(units=256))
    model.add(LeakyReLU(alpha=0.2))
    
    model.add(Dense(units=512))
    model.add(LeakyReLU(alpha=0.2))
    
    model.add(Dense(units=3136, activation='tanh'))   #784層輸出
    
    model.compile(loss='binary_crossentropy', optimizer=Adam(0.0002,0.5))
    return model

generator = build_generator()
generator.summary()

def build_discriminator():
    model = Sequential()
    
    model.add(Dense(units=1024,input_dim=3136))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dropout(0.3))
    
    model.add(Dense(units=512))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dropout(0.3))
    
    model.add(Dense(units=256))
    model.add(LeakyReLU(alpha=0.2))
    model.add(Dropout(0.3))
    
    model.add(Dense(units=1,activation='sigmoid')) #輸出一個
    
    model.compile(loss='binary_crossentropy', optimizer=Adam(0.0002,0.5))
    return model

discriminator = build_discriminator()
discriminator.summary()

def build_GAN(discriminator,generator):  #把generater和discreminator結合
    discriminator.trainable=False
    GAN_input = Input(shape=(3136,))
    x=generator(GAN_input)
    GAN_output=discriminator(x)
    GAN=Model(inputs=GAN_input, outputs=GAN_output)
    GAN.compile(loss='binary_crossentropy', optimizer=Adam(0.0002,0.5))
    return GAN

GAN = build_GAN(discriminator, generator)
GAN.summary()

def draw_images(generator, epoch, examples=25, dim=(5,5), figsize=(10,10)): #將隨機產生的照片秀出
    noise = np.random.normal(loc=0, scale=1, size=[examples,3136])
    generated_images = generator.predict(noise)
    generated_images = generated_images.reshape(25,56,56)
    plt.figure(figsize=figsize)
    for i in range(generated_images.shape[0]):
        plt.subplot(dim[0],dim[1],i+1)
        reverse_generated_images = []
        for j in generated_images[i]:
            reverse_generated_images.append(-j)
        plt.imshow(reverse_generated_images,interpolation='nearest',cmap='Greys')
        plt.axis('off')
    plt.tight_layout()
    plt.savefig('Generated_images %d.png' %epoch)


def draw_images_2(generator, examples=25, dim=(1,1), figsize=(10,10)): #將隨機產生的照片秀出
#     noise = np.random.normal(loc=0, scale=1, size=[examples,784])

    #測試照片
    img = Image.open('C:/Users/asus/children_test/test6.jpg')
    reIm = img.resize((56,56))
    im1 = np.array(reIm.convert("L"))   #將圖檔轉為陣列儲存，L轉為灰階陣列

    plt.imshow(im1 , cmap=plt.get_cmap('gray'))
    plt.show

    im1 = im1.reshape(1,56*56)   #將28*28轉乘1*784
    im1 = im1.astype('float32')/255
#     #顛倒黑白 正規畫
    im1 -= 0.5
    im1 *= -1
    im1 *= 7
    
    noise = [im1]
    noise = np.array(noise)
    generated_images = generator.predict(noise)
    generated_images = generated_images.reshape(1,56,56)
    plt.figure(figsize=figsize)
    for i in range(generated_images.shape[0]):
        plt.subplot(dim[0],dim[1],i+1)
        reverse_generated_images = []
        for j in generated_images[i]:
            reverse_generated_images.append(-j)
        plt.imshow(reverse_generated_images,interpolation='nearest',cmap='Greys')
        plt.axis('off')
    plt.tight_layout()
    plt.savefig('Children_images .png')



def train_GAN(epochs=500, batch_size=128): 
    
    #Loading the data
    x_train,y_train=load_data()
    
    generator=build_generator()
    discriminator=build_discriminator()
    GAN=build_GAN(discriminator, generator)

    for i in range(1, epochs+1):
        print("Epoch %d" %i)
        
        
        for _ in tqdm(range(batch_size)):   #每次都是128張照片訓練
            noise = np.random.normal(0, 1,(batch_size, 3136))
            fake_images= generator.predict(noise)
           
            real_images=x_train[np.random.randint(0,x_train.shape[0], batch_size)]

            label_fake = np.zeros(batch_size)
            label_real = np.ones(batch_size)
            
            x=np.concatenate([fake_images, real_images])
            y=np.concatenate([label_fake, label_real])
            
            discriminator.trainable=True
            discriminator.train_on_batch(x,y)
            
            discriminator.trainable=False
            GAN.train_on_batch(noise, label_real)
        if i==1 or i%2==0:
            draw_images(generator,i) 
    
    draw_images_2(generator)
            
train_GAN(epochs=40, batch_size=128)          
```
